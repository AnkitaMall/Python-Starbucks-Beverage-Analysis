# **Import Libraries**

```python

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import os
for dirname, _, filenames in os.walk('/kaggle/input/starbucks'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

# **Reading CSV**

```python
df = pd.read_csv('/kaggle/input/starbucks/starbucks.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['Beverage_category', 'Trans Fat (g) ', 'Saturated Fat (g)', ' Sodium (mg)', ' Total Carbohydrates (g) ', ' Dietary Fibre (g)', 'Vitamin A (% DV) ', 'Vitamin C (% DV)', ' Calcium (% DV) ', 'Iron (% DV) '], axis=1)

df.head()
```

```python
df.info
```

# **Cleaning the data**

# dropping duplicates

```python
df.drop_duplicates(inplace= True)
df.shape
```

# count plot of various beverage consume

```python
plt.figure(figsize=(16, 10))
ax = sns.countplot(data=df, x='Beverage')
plt.xticks(rotation=90)
plt.show()
```

# **Visualisation**

# distribution of calories

```python

plt.figure(figsize=(10, 6))
plt.hist(df['Calories'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Calories Across Beverages')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Sugare content in beverage

```python
plt.figure(figsize=(12, 8))
plt.bar(df['Beverage'], df[' Sugars (g)'], color='lightcoral', edgecolor='black')
plt.title('Sugar Content Across Different Beverages')
plt.xlabel('Beverage')
plt.ylabel('Sugar Content (grams)')
plt.xticks(rotation=90, ha='right')  
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Protein distribution

```python
plt.figure(figsize=(12, 8))
plt.bar(df['Beverage'], df[' Protein (g) '], color='lightgreen', edgecolor='black')
plt.title('Protein Content Across Different Beverages')
plt.xlabel('Beverage')
plt.ylabel('Protein Content (grams)')
plt.xticks(rotation=90, ha='right') 
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
